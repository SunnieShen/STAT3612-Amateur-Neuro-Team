```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 假设 df 是已有的 DataFrame（从 example.csv 加载）
# 如果没有数据，可以用随机数据示例
np.random.seed(42)
df = pd.DataFrame({
    'one': np.random.randn(1000),
    'two': np.random.randn(1000),
    'three': np.random.randn(1000),
    'four': np.random.randn(1000)
})

# ================== 图1: 散点图 ==================
plt.figure(figsize=(6, 5))
plt.scatter(df.iloc[:100, 1], df.iloc[:100, 2], alpha=0.7, edgecolors='k')
plt.xlabel('two')
plt.ylabel('three')
plt.title('Scatter Plot: two vs three (first 100 samples)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ================== 图2: 直方图 ==================
plt.figure(figsize=(6, 5))
plt.hist(df.iloc[:, 3], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('four')
plt.ylabel('Frequency')
plt.title('Histogram of column "four"')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ================== 图3: 单条折线图 ==================
plt.figure(figsize=(8, 5))
plt.plot(df.iloc[:100, 1], color='red', linewidth=2, marker='o', markersize=4)
plt.xlabel('Index (first 100 samples)')
plt.ylabel('Value of "two"')
plt.title('Line Plot of column "two"')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ================== 图4: 多条折线图（子图） ==================
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Multiple Line Plots', fontsize=16)

# 子图1
axes[0, 0].plot(df.iloc[:50, 0], color='blue', label='one')
axes[0, 0].set_title('Column: one')
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.5)

# 子图2
axes[0, 1].plot(df.iloc[:50, 1], color='red', label='two')
axes[0, 1].set_title('Column: two')
axes[0, 1].legend()
axes[0, 1].grid(True, linestyle='--', alpha=0.5)

# 子图3
axes[1, 0].plot(df.iloc[:50, 2], color='green', label='three')
axes[1, 0].set_title('Column: three')
axes[1, 0].legend()
axes[1, 0].grid(True, linestyle='--', alpha=0.5)

# 子图4
axes[1, 1].plot(df.iloc[:50, 3], color='orange', label='four')
axes[1, 1].set_title('Column: four')
axes[1, 1].legend()
axes[1, 1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
